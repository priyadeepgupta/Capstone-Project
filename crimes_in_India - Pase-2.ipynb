```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

crime_data_url = 'https://github.com/priyadeepgupta/capstone/raw/main/Crime_Data.xlsx'

crime_df = pd.read_excel(crime_data_url, sheet_name='Crime') 
literacy_df = pd.read_excel(crime_data_url, sheet_name='Literacy')  

merged_df = pd.merge(crime_df, literacy_df, on='State_UT')


plt.figure(figsize=(12, 6))


sns.barplot(x='State_UT', y='Crime_Rate', data=merged_df, palette='viridis')

for index, row in merged_df.iterrows():
    plt.text(index, row['Crime_Rate'], f'{row["Literacy Rate"]:.2f}', ha='center', va='bottom')


plt.title('Total Crimes by State with Literacy Rate Annotation')
plt.xlabel('State')
plt.ylabel('Crime_Rate')

plt.xticks(rotation=90)


plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

crime_data_url = 'https://github.com/priyadeepgupta/capstone/raw/main/Crime_Data.xlsx'


crime_type_df = pd.read_excel(crime_data_url, sheet_name='Crime_type')
literacy_df = pd.read_excel(crime_data_url, sheet_name='Literacy')

merged_df = pd.merge(crime_type_df, literacy_df, on='State_UT')

merged_df.set_index('State_UT', inplace=True)

crime_types = ['Violent Crimes', 'Murder', 'Rape', 'Kidnapping','Crimes Against Children','Extortion','Robbery','Hit & Run','Drugs Trafficking','Illegal Arms']


for crime_type in crime_types:
   
    plt.figure(figsize=(12, 6))

    sns.lineplot(data=merged_df, x=merged_df.index, y=crime_type, color='blue', label=crime_type)
    
    sns.lineplot(data=merged_df, x=merged_df.index, y='Literacy Rate', color='red', label='Literacy Rate')

    for i, value in enumerate(merged_df[crime_type]):
        plt.text(i, value, f'{value:.2f}', va='bottom', ha='center', fontsize=8, rotation=90, color='blue')

    for i, value in enumerate(merged_df['Literacy Rate']):
        plt.text(i, value, f'{value:.2f}', va='bottom', ha='center', fontsize=8, rotation=90, color='red')

    plt.title(f'{crime_type} vs Literacy Rate in Each State')
    plt.xlabel('State')
    plt.ylabel('Rate')

    plt.xticks(rotation=90)

    plt.legend()

    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

crimedata_url = 'https://github.com/priyadeepgupta/capstone/raw/main/Statewise_crimes_committed.xlsx'

data = pd.read_excel(crimedata_url)

yearly_crime_rate = data.groupby('YEAR')['TOTAL IPC CRIMES'].sum()

plt.figure(figsize=(10, 6))
yearly_crime_rate.plot(marker='o', color='b', linestyle='-')
plt.title('Year-on-Year Total Crime Rate')
plt.xlabel('Year')
plt.ylabel('Total Crime Rate')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

crime_data_url = 'https://github.com/priyadeepgupta/capstone/raw/main/Crime_Data.xlsx'
try:
    crime_df = pd.read_excel(crime_data_url, sheet_name='Crime_Type_Total')
    population_df = pd.read_excel(crime_data_url, sheet_name='Population')
except FileNotFoundError:
    print("File not found. Please check the file path.")

crime_df.rename(columns={'STATE_UT': 'State_UT'}, inplace=True)

try:
    merged_df = pd.merge(crime_df, population_df, on='State_UT')
except KeyError as e:
    print(f"KeyError: {e}. Please check if the 'State_UT' column exists in both DataFrames.")

crime_types = ["MURDER", "ATTEMPT TO MURDER", "CULPABLE HOMICIDE NOT AMOUNTING TO MURDER", "RAPE",
               "CUSTODIAL RAPE", "OTHER RAPE", "KIDNAPPING & ABDUCTION",
               "KIDNAPPING AND ABDUCTION OF WOMEN AND GIRLS", "KIDNAPPING AND ABDUCTION OF OTHERS",
               "DACOITY", "PREPARATION AND ASSEMBLY FOR DACOITY", "ROBBERY", "BURGLARY", "THEFT",
               "AUTO THEFT", "OTHER THEFT", "RIOTS", "CRIMINAL BREACH OF TRUST", "CHEATING",
               "COUNTERFIETING", "ARSON", "HURT/GREVIOUS HURT", "DOWRY DEATHS",
               "ASSAULT ON WOMEN WITH INTENT TO OUTRAGE HER MODESTY", "INSULT TO MODESTY OF WOMEN",
               "CRUELTY BY HUSBAND OR HIS RELATIVES", "IMPORTATION OF GIRLS FROM FOREIGN COUNTRIES",
               "CAUSING DEATH BY NEGLIGENCE", "OTHER IPC CRIMES"]
merged_df['Total_Crime'] = merged_df[crime_types].sum(axis=1)

merged_df['Overall Crime Rate Per Capita'] = (merged_df['Total_Crime'] / merged_df['Total_Population']) * 1000

plt.figure(figsize=(10, 6))
sns.scatterplot(data=merged_df, x='Total_Population', y='Overall Crime Rate Per Capita', color='blue')
plt.title('Relationship Between Population and Overall Crime Rate Per Capita')
plt.xlabel('Population')
plt.ylabel('Overall Crime Rate Per Capita (per 1000 people)')
plt.grid(True)
plt.show()
```

# State-wise Crime Report

Introduction: This report presents an analysis of crime statistics across different states and union territories (UTs) in India for the year 2021. The data includes total reported crimes, crime rates, crime density, investigation rates, charge-sheeting rates, and conviction rates. The analysis aims to provide insights into the crime scenario and law enforcement effectiveness in various regions of the country.

Key Insights:
1.	Andhra Pradesh: With a total of 222,199 reported crimes in 2021, Andhra Pradesh experienced a crime rate of 420.4 per 100,000 population. The state demonstrated a relatively high investigation rate of 63.7% and a commendable charge-sheeting rate of 92.9%. However, the conviction rate of 84.7% indicates room for improvement in legal outcomes.
2.	Maharashtra: Maharashtra reported 540,800 crimes, resulting in a crime rate of 433.5 per 100,000 population. The state exhibited moderate investigation and charge-sheeting rates of 62.1% and 76.4%, respectively. However, the conviction rate of 54.4% suggests challenges in securing legal convictions.
3.	Kerala: Kerala recorded a staggering 524,960 crimes in 2021, translating to a high crime rate of 1477.2 per 100,000 population. Despite the high crime prevalence, Kerala demonstrated impressive investigation, charge-sheeting, and conviction rates of 84.9%, 91.8%, and 86.5%, respectively. This indicates robust law enforcement mechanisms in the state.
4.	Uttar Pradesh: Uttar Pradesh reported the highest number of crimes in India, with 608,082 cases and a crime rate of 262.4 per 100,000 population. The state exhibited a relatively high investigation rate of 86% but lower charge-sheeting and conviction rates of 75.5% and 63.4%, respectively. This suggests a need for more effective legal processes to ensure justice.
5.	Delhi: Delhi reported 306,389 crimes, resulting in a remarkably high crime rate of 1479.9 per 100,000 population. The city-state demonstrated moderate investigation and conviction rates of 73.8% and 86.6%, respectively, but a low charge-sheeting rate of 31%. This indicates challenges in the legal process and the need for improved charge-sheeting practices.

Key Findings:
1.	Overall Crime Rate: India reported a total of 6,096,310 crimes in 2021, with a national crime rate of 445.9 per 100,000 population. Crime rates varied significantly among states and UTs, influenced by factors such as population density, socio-economic conditions, and law enforcement capabilities.
2.	Crime Density: Crime density, measured as the number of crimes per 100 square kilometers, ranged from as low as 3.62 in Arunachal Pradesh to as high as 20,660 in Delhi. This indicates wide disparities in crime concentration across different regions.
3.	Investigation and Conviction Rates: Investigation, charge-sheeting, and conviction rates varied across states. States like Mizoram, Delhi, and Ladakh demonstrated relatively high investigation, charge-sheeting, and conviction rates, reflecting robust law enforcement mechanisms. Conversely, states like Assam, Odisha, and West Bengal reported lower investigation and conviction rates, suggesting challenges in the criminal justice system.
4.	Crime Trends: Some states exhibited consistent trends across multiple parameters. For example, Kerala reported a high overall crime rate, crime density, and relatively high investigation and conviction rates, indicating both high crime prevalence and effective law enforcement efforts. On the other hand, states like Manipur and Nagaland reported relatively low crime rates but also low investigation and conviction rates, indicating potential gaps in law enforcement.
5.	Urban vs. Rural Dynamics: Urban areas generally reported higher crime rates and crime densities compared to rural areas. However, the effectiveness of law enforcement varied within urban and rural regions, with some urban centre’s demonstrating better investigation and conviction rates despite higher crime rates.

Insights and Recommendations:
1.	Enhanced Law Enforcement Strategies: States with high crime rates should prioritize the implementation of proactive law enforcement strategies, including increased police presence, community policing initiatives, and intelligence-led operations to deter criminal activities.
2.	Strengthening Legal Processes: Improving investigation, charge-sheeting, and conviction rates is crucial for ensuring justice and deterring offenders. States with lower legal outcomes should invest in capacity building for law enforcement personnel, modernization of forensic facilities, and expedited trial processes.
3.	Addressing Socio-Economic Factors: Many crimes stem from socio-economic disparities, unemployment, and lack of educational opportunities. States must adopt holistic approaches that address root causes of crime, such as poverty alleviation programs, skill development initiatives, and youth engagement activities.
4.	Community Engagement and Awareness: Building trust and collaboration between law enforcement agencies and communities is essential for effective crime prevention. States should prioritize community engagement initiatives, public awareness campaigns, and crime prevention programs tailored to local needs.
5.	Investment in Technology and Infrastructure: Leveraging technology, such as CCTV surveillance, data analytics, and crime mapping tools, can enhance law enforcement capabilities and facilitate targeted interventions. States should invest in upgrading infrastructure and adopting innovative technologies to combat crime effectively.

Conclusion:
The analysis reveals significant variations in crime prevalence and law enforcement effectiveness across different states and UTs in India. While some states demonstrate high crime rates but effective law enforcement mechanisms (e.g., Kerala), others face challenges in both crime control and legal outcomes (e.g., Uttar Pradesh). Policymakers and law enforcement agencies must tailor interventions to address the specific needs and challenges of each region, focusing on enhancing investigation, charge-sheeting, and conviction processes to ensure justice and maintain law and order. Additionally, efforts should be made to address underlying socio-economic factors contributing to crime and strengthen community policing initiatives to foster trust and cooperation between law enforcement agencies and the public.
